```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the reads seen 
refs_file = 'new_1_11_3.refs.txt'
oligos_read = []
with open(refs_file) as f:
    for line in f:
        oligos_read.append(int(line.rstrip('\n').split('_')[-1]))
oligos_read = np.array(oligos_read)
```

```python
coverage = np.zeros(734,dtype=int)
for oligo in oligos_read:
    coverage[oligo] += 1
```

```python
plt.plot(coverage)
plt.xlabel('oligo#')
plt.ylabel('coverage')
```

```python
# now load gc biases of the oligos
path = '/raid/nanopore/shubham/20200214_nanopore_pool_data/data/oligo_files/oligos_2.fa'
gc_fraction = np.zeros(734)
oligo_num = 0
even = True
with open(path) as f:
    for line in f:
        if even:
            even = False
            continue
        even = True
        line = line.rstrip('\n')
        gc_fraction[oligo_num] = (line.count('G')+line.count('C'))/len(line)
        oligo_num += 1
```

```python
plt.figure()
plt.plot(gc_fraction)
```

```python
plt.hist(gc_fraction,histtype='step',bins=15)
```

```python
plt.hist(gc_fraction,histtype='step',bins=15,cumulative=True,density=True)
plt.grid()
```

```python
plt.scatter(gc_fraction,coverage,alpha=0.1)
plt.xlabel('GC fraction')
plt.ylabel('Coverage')
```

```python
import scipy.stats
```

```python
scipy.stats.spearmanr(gc_fraction,coverage)
```

```python
scipy.stats.spearmanr(np.abs(0.5-gc_fraction),coverage)
```

```python
scipy.stats.spearmanr(gc_fraction,coverage<500)
```

```python
scipy.stats.pearsonr(gc_fraction,coverage)
```

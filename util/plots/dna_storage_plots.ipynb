```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```
data = [(0.86,3.04,"This work","Genapsys"),
        (0.73,4.29,"This work","Genapsys"),
        (0.91,2.73,"Chandak et al. (2019)","Illumina"),
        (0.67,3.82,"Chandak et al. (2019)","Illumina"),
        (0.91,4.55,"Organick et al. (2018)","Illumina"),
        (0.65,6.80,"Erlich and Zielinski (2017)","Illumina"),
        (0.93,7.01,"Chandak et al. (2020)","Nanopore"),
        (1.85,4.42,"Chandak et al. (2020)","Nanopore"),
        (1.07,6.59,"Chandak et al. (2020)","Nanopore"),
        (0.91,20.02,"Lopez et al. (2019)","Nanopore"),
        (0.85,3.23,"This work","Nanopore"),
        (0.88,3.07,"This work","Nanopore"),
        (1.02,2.89,"This work","Nanopore"),
]
```

```
color_map = {'Genapsys':'r','Illumina':'b','Nanopore':'darkgreen'}
marker_map = {'This work':'*',"Chandak et al. (2019)":'2',\
              "Chandak et al. (2020)":'3',\
              "Erlich and Zielinski (2017)":'>',\
              "Lopez et al. (2019)":"^",\
              "Organick et al. (2018)":'<',\
              }
fig, ax = plt.subplots(figsize=(8,5))
color_legend_elements = [Line2D([0], [0], marker='o', color='w', label=k,
                          markerfacecolor=color_map[k], markersize=10) for k in color_map]
marker_legend_elements = [Line2D([0], [0], marker=marker_map[k], color='w', label=k,
                          markeredgecolor='k', markerfacecolor='k', markersize=10) for k in marker_map]
for data_point in data:
  plt.scatter(data_point[1],data_point[0],\
              marker=marker_map[data_point[2]],\
              s=80,
              c=color_map[data_point[3]])

leg1 = ax.legend(handles=color_legend_elements,loc="upper left",bbox_to_anchor=(1.0,1))
leg2 = ax.legend(handles=marker_legend_elements,loc="lower left",bbox_to_anchor=(1.0,0))
ax.add_artist(leg1)
ax.grid(b=True, which='major')
ax.grid(b=True, which='minor',alpha=0.2)
ax.minorticks_on()
ax.set_xlim(left=0.0)
ax.set_ylim(bottom=0.5)
ax.set_xlabel('Reading cost (bases/bit)')
ax.set_ylabel('Writing cost (bases/bit)')
plt.savefig('main.png',bbox_inches='tight')
```

```
color_map = {'6':'r','8':'b','11':'darkgreen'}
marker_map = {'3/4':'1','5/6':'2','7/8':'3'}
fig, ax = plt.subplots(figsize=(8,5))
color_legend_elements = [Line2D([0], [0], marker='o', color='w', label='Memory: '+k,
                          markerfacecolor=color_map[k], markersize=10) for k in color_map]
marker_legend_elements = [Line2D([0], [0], marker=marker_map[k], color='w', label='Rate: '+k,
                          markeredgecolor='k', markerfacecolor='k', markersize=10) for k in marker_map]
for data_point in data_m_r_1crc:
  ax.scatter(data_point[3],data_point[2],\
              marker=marker_map[data_point[1]],\
              s=80,
              c=color_map[data_point[0]])
  ax.annotate(s="("+data_point[0]+","+data_point[1]+")",\
              xy=(data_point[3],data_point[2]),\
              textcoords="offset points",\
              xytext=(5,0),
              ha="left")

leg1 = ax.legend(handles=color_legend_elements,loc="upper left",bbox_to_anchor=(1.0,1))
leg2 = ax.legend(handles=marker_legend_elements,loc="lower left",bbox_to_anchor=(1.0,0))
ax.add_artist(leg1)
ax.grid(b=True, which='major')
ax.grid(b=True, which='minor',alpha=0.2)
ax.minorticks_on()
ax.set_xlim(left=2.5)
ax.set_ylim(bottom=0.8)
ax.set_xlabel('Reading cost (bases/bit)')
ax.set_ylabel('Writing cost (bases/bit)')
plt.savefig('conv_m_r.png',bbox_inches='tight')
```

```
data_m_r_2crc = [#('6',"3/4",1.05,0.00),\
                 ('8',"3/4",1.07,3.53),\
                 ('11',"3/4",1.09,3.87),\
                 ('6',"5/6",0.91,4.62),\
                 ('8',"5/6",0.92,3.49),\
                 ('11',"5/6",0.94,3.49),\
                 ('6',"7/8",0.87,3.56),\
                 ('8',"7/8",0.88,3.87),\
                 ('11',"7/8",0.89,3.37)
                 ]
```

```
color_map = {'6':'r','8':'b','11':'darkgreen'}
marker_map_1crc = {'3/4':'1','5/6':'2','7/8':'3'}
marker_map_2crc = {'3/4':'v','5/6':'^','7/8':'<'}
fig, ax = plt.subplots(figsize=(8,5))
color_legend_elements = [Line2D([0], [0], marker='o', color='w', label='Memory: '+k,
                          markerfacecolor=color_map[k], markersize=10) for k in color_map]
marker_legend_elements = [Line2D([0], [0], marker=marker_map_1crc[k], color='w', label='Rate: '+k+", 1 CRC",
                          markeredgecolor='k', markerfacecolor='k', markersize=10)\
                          for k in marker_map_1crc] + \
                         [Line2D([0], [0], marker=marker_map_2crc[k], color='w', label='Rate: '+k+", 2 CRC",
                          markeredgecolor='k', markerfacecolor='k', markersize=10)\
                          for k in marker_map_2crc]
for data_point in data_m_r_1crc:
  ax.scatter(data_point[3],data_point[2],\
              marker=marker_map_1crc[data_point[1]],\
              s=80,
              c=color_map[data_point[0]])
for data_point in data_m_r_2crc:
  ax.scatter(data_point[3],data_point[2],\
              marker=marker_map_2crc[data_point[1]],\
              s=80,
              c=color_map[data_point[0]])
  ax.annotate(s="("+data_point[0]+","+data_point[1]+")",\
              xy=(data_point[3],data_point[2]),\
              textcoords="offset points",\
              xytext=(5,0),
              ha="left")

data_map_1crc = {(data_point[0],data_point[1]):(data_point[2],data_point[3]) for data_point in data_m_r_1crc}
data_map_2crc = {(data_point[0],data_point[1]):(data_point[2],data_point[3]) for data_point in data_m_r_2crc}

# draw lines connecting corresponding points
for (m,r) in data_map_1crc:
  if (m,r) in data_map_2crc:
    startx = data_map_1crc[(m,r)][1]
    starty = data_map_1crc[(m,r)][0]
    endx = data_map_2crc[(m,r)][1]
    endy = data_map_2crc[(m,r)][0]
    plt.plot([startx,endx],[starty,endy],c=color_map[m],ls='dotted',alpha=0.5)

leg1 = ax.legend(handles=color_legend_elements,loc="upper left",bbox_to_anchor=(1.0,1))
leg2 = ax.legend(handles=marker_legend_elements,loc="lower left",bbox_to_anchor=(1.0,0))
ax.add_artist(leg1)
ax.grid(b=True, which='major')
ax.grid(b=True, which='minor',alpha=0.2)
ax.minorticks_on()
ax.set_xlim(left=2.5)
ax.set_ylim(bottom=0.8)
ax.set_xlabel('Reading cost (bases/bit)')
ax.set_ylabel('Writing cost (bases/bit)')
plt.savefig('conv_2crc.png',bbox_inches='tight')
```
